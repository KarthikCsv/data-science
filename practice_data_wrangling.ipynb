```python
#import statement

import pandas as pd
import matplotlib.pylab as plt
```

```python
file = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(file, names = headers)
```

```python
df.head()
```

```python
import numpy as np

# replace '?' as NaN

df.replace("?", np.nan, inplace = True)
```

```python
df.head()
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_norm_loss = df["normalized-losses"].astype(float).mean(axis=0)
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
print(avg_norm_loss)
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_stroke = df["stroke"].astype(float).mean(axis=0)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_bore = df["bore"].astype(float).mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
avg_hp = df["horsepower"].astype(float).mean(axis=0)
df["horsepower"].replace(np.nan, avg_hp, inplace = True)
```

```python
avg_rpm = df["peak-rpm"].astype(float).mean(axis=0)
df["peak-rpm"].replace(np.nan, avg_rpm, inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, "four", inplace = True)
```

```python
#to drop prices of prices
df.dropna(subset=['price'], axis = 0, inplace = True)
```

```python
#reset index

df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
missing_data = df.isnull()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)

df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
# Convert highway-mpg to highway-L/100km by mathematical operation (235 divided by mpg)

df['highway-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
```

```python
df.head()
```

```python
#normalization for length height and width

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
df[["length","width","height"]].head()
```

```python
df['horsepower'] = df['horsepower'].astype(int, copy = True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

#set x y labels as plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower_binned'] = pd.cut(df['horsepower'], bins, labels = group_names, include_lowest = True)
df[['horsepower','horsepower_binned']].head(20)
```

```python
dummy_variable1 = pd.get_dummies(df['fuel-type'])
dummy_variable1.head()
```

```python
dummy_variable1.rename(columns = {'fuel-type-diesel': 'diesel', 'fuel-type-gas' : 'gas'}, inplace = True)
dummy_variable1.head()
```

```python
#merge data frame 'df' and 'dummy variable'

df = pd.concat([df, dummy_variable1], axis =1)
df.drop('fuel-type', axis = 1, inplace = True)
```

```python
df.head()
```

```python
dummy_variable2 = pd.get_dummies(df['aspiration'])
dummy_variable2.head()
```

```python
dummy_variable2.rename(columns = {'aspiration-std' : 'std', "aspiration-turbo" : "turbo"}, inplace = True)
dummy_variable2.head()
```

```python
df = pd.concat([df, dummy_variable2], axis =1)
df.drop('aspiration', axis = 1, inplace =True)
```
